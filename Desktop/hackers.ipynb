```python
ask_posts,show_posts,other_posts= [],[],[]
for i in hn:
    title = i[1]
    check = title.lower()
    if check.startswith('ask hn'):
        ask_posts.append(i)
    elif check.startswith('show hn'):
        show_posts.append(i)
    else:
        other_posts.append(i)
print(len(ask_posts),len(show_posts),len(other_posts))
```

```python
total_ask_comments = 0
total = 0
for i in ask_posts:
    comments = int(i[4])
    total_ask_comments += comments
    total+=1
avg_ask_comments = total_ask_comments/total 
avg_ask_comments
```

```python
total_show_comments = 0
totals = 0
for i in show_posts:
    comments = int(i[4])
    total_show_comments += comments
    totals +=1
avg_show_comments = total_show_comments / totals
avg_show_comments
```

We can see on an average ask posts get more comments than show posts
The remaining analysis we'll continue in ask_posts

we'll determine if ask posts created at a certain time are more likely to attract comments.

```python
import datetime as dt 
result_list  = []
for i in ask_posts:
    time = i[6]
    comments = int(i[4])
    result_list.append([time, comments])
```

```python
result_list[:2]
```

Process of extracting hour

```python
x= '8/16/2016 9:55'
print(dt.datetime.strptime(x,'%m/%d/%Y %H:%M').strftime('%H'))
```

```python
counts_by_hour, comments_by_hour = {},{}
dateformat = '%m/%d/%Y %H:%M'
for i in result_list:
    date  =  i[0].replace('-','/')
    time = dt.datetime.strptime(date,dateformat).strftime('%H')
    comments = i[1]
    if time in counts_by_hour:
        counts_by_hour[time] += 1
        comments_by_hour[time] += comments 
    else:
        counts_by_hour[time] = 1
        comments_by_hour[time] = comments 
comments_by_hour
```

Show the dict in key sorted manner

```python
for i in sorted(comments_by_hour):
    print(i,'->', comments_by_hour[i])
```

In Value sorted

```python
sorted_item = sorted(comments_by_hour.items(),key = lambda item:item[1], reverse=True)
for i in sorted_item:
    print(i[0],'->',i[1])
```

```python
counts_by_hour
```

Calculate the average number of comments per post for posts created during each hour of the day

```python
avg_hours = []
for i in comments_by_hour.keys():
    avg = comments_by_hour[i]/counts_by_hour[i]
    avg_hours.append([i,avg])
avg_hours = sorted(avg_hours)
avg_hours
```

Sort by avg descending

```python
sorted_avg_hours = sorted(avg_hours, key = lambda item: item[1], reverse = True )
sorted_avg_hours
```

```python
# or ,

for i in sorted_avg_hours:
    print(i[1],'->', i[0])
```

Top 5 Hours for Ask Posts Comments

```python
for i in sorted_avg_hours[:5]:
    print(dt.datetime.strptime(i[0],'%H').strftime('%H:%M'),
          '->', "{:.2f}".format(i[1]),'average comments per post')
```

The hour that receives the most comments per post on average is 15:00, with an average of 38.59 comments per post
