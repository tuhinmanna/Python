Summarizing Appointment Lengths

```python
for row in potus:
    end_date = row[3]
    end_date = dt.datetime.strptime(end_date, "%m/%d/%y %H:%M")
    row[3] = end_date
    
appt_lengths = {}
for i in potus:
    start = i[2]
    finish = i[3]
    length = finish - start 
    if length in appt_lengths:
        appt_lengths[length] += 1
    else:
        appt_lengths[length] = 1
        
min_length = min(appt_lengths)
max_length = max(appt_lengths)
```
