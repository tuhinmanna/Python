## Remove non-english apps

All the english letter have an ASCII value of less than 127.

````python
def english(string):
    for i in string:
        if(ord(i)>127):
            print("Non English")
            return
    print('english')
````

````python
english('abcdAA')
english('‡§π‡§ø‡§Ç‡§¶‡•Ä ‡§ü‡§æ‡§á‡§™‡§ø‡§Ç‡§ó')
english('abcde@@@@ü§ó')
````

However, english apps can have emojis as well, which is having a higher ASCII  value

````python
# So, we'll consider if an appname have more than 3 such characters , we'll reject that app
````

````python
def new_english(string):
    counter = 0
    for i in string:
        if ord(i)>127:
            counter+=1
    if counter>3:
        return False
    else:
        return True
````

Creating frquency table

````python
def freq_table(dataset, index):
    table = {}
    total = 0
    for i in dataset:
        total+=1
        if i[index] in table:
            table[i[index]]+=1
        else:
            table[i[index]] = 1
            
    table_perc = {}
    for x in table:
        percentage = (table[x] / total) * 100
        table_perc[x] = percentage
    return table_perc
    
````

````python
# First we are creating a freq table, but since later we need the % , we are initiating a total value variable
# Creating a seperate dict to store this percentage wise frequency table
````

````python
def display_table(dataset, index):
    d_table = freq_table(dataset, index)
    table_list = []
    for i in d_table:
        new_tuple = (d_table[i], i)
        table_list.append(new_tuple)
    sorted_list = sorted(table_list, reverse = True)
    for val in sorted_list:
        print(val[1] , ':' , val[0])
````

Now, we'd like to determine the kind of apps with the most users. <br>
One way to find out what genres are the most popular (have the most users) is to calculate the average <u>number of installs</u> for each app genre. For the Google Play data set, we can find this information in the <b>Installs</b> column, but this information is missing for the App Store data set. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the <b>rating_count_tot</b> column

Calculating the average number of user ratings per app genre on the App Store

## Most Popular app on AppStore by Genre

````python
dict = freq_table(ios_final, -5)
free_list = []
for genre in dict:
    total = 0
    rating = 0
    for i in ios_final:
        if i[-5] == genre:
            total+=1
            f = float(i[5])
            rating+=f
    avg = rating/total
    new_tuple = (avg,genre)
    free_list.append(new_tuple)
    #print(genre, ':', avg) 

f = sorted(free_list, reverse = True)
for z in f:
    print(z[1] ,':', z[0])
````

````python
# This has another easy way of calculation in o(n) complexity, the above code is having complexity o(n^2)
````

````python
n_dict = {}
for i in ios_final:
    if i[-5] not in n_dict:
        n_dict[i[-5]] = [0,0]
    
    n_dict[i[-5]][0]+= float(i[5])
    n_dict[i[-5]][1]+=1
    
f_list = []
for genre, (rating,total) in n_dict.items():
    avg=rating/total
    f_list.append((avg,genre))
t_tuple = sorted(f_list, reverse = True)
for x in t_tuple:
    print(x[1], ':' , x[0])
````

## Google play dataset analysis


````python
g_play = {}
g_list = []
for i in android_final:
    
    cat = i[1]
    inst = i[5]
    if cat not in g_play:
        g_play[cat] = [0,0]
    inst = inst.replace(',' , '')
    inst = inst.replace('+' , '')
    g_play[cat][0]+=int(inst)
    g_play[cat][1]+=1
for category, (total_install, total_apps) in g_play.items():
    avg = total_install/total_apps
    g_list.append([avg, category])

new_list = sorted(g_list, reverse =True)
for i in new_list:
    print(i[1], ':' , i[0])
````

On average, communication apps have the most installs: 38,456,119. This number is heavily skewed up by a few apps that have over one billion installs (WhatsApp, Facebook Messenger, Skype, Google Chrome, Gmail, and Hangouts), and a few others with over 100 and 500 million installs:
